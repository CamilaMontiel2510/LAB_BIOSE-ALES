```python
def inspeccion_sujetos(h):
        #punto 1
    t1 = np.arange(0,4+0.1, 0.1)
    t2 = np.arange(0,10+0.2,0.2)

    s1 = 5*np.sin(2*np.pi*0.5*t1)
    s2 = np.sin(2*np.pi*0.25*t2)
    #matriz
    matriz1 = np.array([t1,s1])
    matriz2 = np.array([t2,s2])
    #punto 2: graficación
    plt.figure(figsize = (10,6))
    plt.subplot(1,2,1)
    plt.plot(matriz1[0], matriz1[1], linestyle='--', marker='o', color='r', linewidth=2)
    plt.title('Señal Sujeto 1')
    plt.xlabel('Tiempo(s)')
    plt.ylabel('Amplitud')

    plt.subplot(1,2,2)
    plt.plot(matriz2[0], matriz2[1], linestyle=':', marker='X', color='b', linewidth=2)
    plt.title('Señal Sujeto 2')
    plt.xlabel('Tiempo(s)')
    plt.ylabel('Amplitud')
    #valor de variable B
    B = np.sum(s1*h)/h

    cantidad_s2_m= []
    for i in range(len(s2)):
        if s2[i] < h:
            cantidad_s2_m.append(s2[i])
    info = len(cantidad_s2_m)
    data_= ''
    if info % 2 == 0: 
        data_ = 'información del sujeto 2 par'
    else:
        data_ = 'información del sujeto 2 impar'
    data = print(f'{data_}')
    return  data
```

```python
from scipy.integrate import solve_ivp

```

```python
def sys(t,y, m, c):
    F = np.where(t>=5,0.1,0)
    x1, x2 = y
    dx1dt = x2
    dx2dt =(1/m)*F - (c/m)*x2
    return [dx1dt, dx2dt]
```
