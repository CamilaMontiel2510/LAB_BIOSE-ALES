ejercicio con periodo como parámetro

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def revision_insulina_t(T, umbral):
    Ts =0.2
    t = np.arange(0,40+Ts,Ts)
    fs = 1/T
    f_t = []
    for i in t:
        if i <= 20:
            f_t.append(10*np.sin(2*np.pi*fs*i))
        else:
            f_t.append(6*np.sin(2*np.pi*fs*i))
    f_t =np.array(f_t)

    

    #parte b
    var_senoidal = []
    var_tiempo= []
    for i in range(len(f_t)):
        if f_t[i] > umbral:
            var_senoidal.append(f_t[i])
            var_tiempo.append(t[i])
    #parte c: calculo de valor RMS
    RMS = np.sqrt(np.mean(f_t**2))
    #punto D

    y = np.ones(len(f_t))*RMS
    plt.figure(figsize=(10,5))
    plt.plot(t,f_t, label = 'f(t)')
    plt.plot(t,y, 'g', label='Valor RMS')
    plt.scatter(var_tiempo,var_senoidal, marker ='x' , color = 'red', label= 'Valores por encima del umbral')
    plt.legend()
    plt.grid()
    plt.show()
    
```

```python
g_rev_in = revision_insulina_t(5, 0.2)
print(g_rev_in)
```
