```python
import numpy as np
import matplotlib.pyplot as plt

def generar_señal(t):
    # Definición de la señal básica f(t)
    # Pulso negativo de 0 a 1, positivo de 2 a 3
    condiciones = [
        (t >= 0) & (t < 1),
        (t >= 2) & (t < 3)
    ]
    valores = [-2, 2]
    return np.select(condiciones, valores, default=0)

t = np.linspace(-1, 5, 1000)
y = generar_señal(t)

plt.plot(t, y)
plt.grid(True)
plt.title("Señal de Estimulación Cerebral Profunda")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definimos la función matemática del escalón unitario u(t)
def u(t):
    return np.where(t >= 0, 1, 0)

# 2. Definimos la señal base del punto 3 usando los escalones
def f_base(t):
    return -2*u(t) + 2*u(t-1) + 2*u(t-2) - 2*u(t-3)

# 3. Configuramos el tiempo y los parámetros del Punto 4
t = np.linspace(-1, 10, 1000)  # Eje de tiempo de -1 a 10 segundos
N_repeticiones = 3             # Número de veces que existe la onda (tu 'N')
Periodo = 4                    # Cada cuánto se repite el patrón (asumimos 4)

# 4. APLICAMOS LA SUMATORIA (Equivalente al símbolo Sigma de la matemática)
# np.sum va a sumar todas las versiones desplazadas de la señal base
y = np.sum([f_base(t - m * Periodo) for m in range(N_repeticiones)], axis=0)

# 5. Graficamos
plt.plot(t, y)
plt.grid(True)
plt.title("Sumatoria de Señales DBS (Punto 4)")
plt.xlabel("Tiempo (t)")
plt.ylabel("Amplitud")
plt.show()
```
