```python
t = np.arange(0,20+0.3, 0.3)

sen_1 = 0.5*np.sin(2*np.pi*0.25*t)
sen_2 =2*np.sin(2*np.pi*0.2*t)


valores_s1=[]
valores_s2=[]
valores_t = []
umbral = 0
for i in range(len(t)):
     if sen_1[i] > umbral and sen_2[i] > umbral:
        valores_t.append(t[i])
        valores_s1.append(sen_1[i])
        valores_s2.append(sen_2[i])

sen_1_c = np.array(valores_s1)
sen_2_c = np.array(valores_s2)    

plt.plot(t, sen_1, label='Seno 0.5 Hz') 
    
plt.plot(t, sen_2, label='Seno 1 Hz')
plt.plot(valores_t, sen_1_c, 'ro', label='Seno 0.5 Hz > 0')
plt.plot(valores_t, sen_2_c, 'go', label='Seno 1 Hz > 0')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ejemplo de señales (puedes reemplazarlas por tus datos)
vector1 = np.array([1, 2, 3, 4, 5, 6])
vector2 = np.array([2, 3, 5, 7, 8, 9])

# 1. Coeficiente de correlación
corr = np.corrcoef(vector1, vector2)
print("Coeficiente de correlación:")
print(corr)

# 2. Ajuste lineal (regresión)
p = np.polyfit(vector1, vector2, 1)  # grado 1 = recta
y_ajuste = np.polyval(p, vector1)

print("Pendiente:", p[0])
print("Intercepto:", p[1])

# 3. Gráfica nube de puntos + recta de regresión
plt.figure()
plt.scatter(vector1, vector2, label="Datos")
plt.plot(vector1, y_ajuste, label="Regresión lineal")
plt.xlabel("Vector 1")
plt.ylabel("Vector 2")
plt.title("Regresión entre señales")
plt.legend()
plt.show()

# 4. Cálculo del error
N = len(vector1)
j = (1/N) * np.sum(vector1 - vector2)

print("Error j =", j)
```
